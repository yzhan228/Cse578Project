```python
import pandas as pd
df = pd.read_table('C:\\Users\\Ayush Singh\\Downloads\\adult.data', sep=", ", header=None)
```

```python
df
```

```python
df.columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
df
```

```python
df.columns = df.columns.str.replace(' ', '')

for i in df.columns:
    df = df[df[i] != '?']
```

```python
below_50k = df[df["income"] == "<=50K"]
```

```python
below_50k
```

```python
above_50k = df[df["income"] == ">50K"]
```

```python
above_50k
```

# Bar Graph for Sex Variable

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
below_50k["sex"].value_counts().plot(kind="bar", color=['r', 'b'])

plt.subplot(1,2,2)
above_50k["sex"].value_counts().plot(kind='bar', color=['r', 'b'])

plt.tight_layout(pad=0.1, w_pad=7.0, h_pad=0.9)

```

# Charts for Age Variable

```python
import seaborn as sns
import numpy as np
fig = plt.figure(figsize=(20,5))
ax = sns.kdeplot(below_50k["age"], label = '<=50K income level', shade=True)
sns.kdeplot(above_50k["age"], label = '>50K income level', shade=True, ax=ax)
plt.xticks(np.arange(0,100,5))
ax.set_xlabel("Age (years)")
ax.set_ylabel("Density")
plt.title("Density Distribution for Age and Income Level", fontsize=30)
ax.legend(loc="upper right")
```

```python
fig = plt.figure(figsize=(20,7))
ax = sns.countplot(data= df, x='age', hue='income')
ax.set_xlabel("Age (years)")
ax.set_ylabel("Count")
ax.legend(loc="upper right") 
plt.title("Countplot for How Income Varies by Age", fontsize=30)

print('Minimum age: ' + str(min(df["age"])))
print('Maximum age: ' + str(max(df["age"])))
```

# Bar Chart for Education Type Variable

```python
plt.subplot(1,2,1)
below_50k["education"].value_counts().plot(kind="bar", figsize=(14,8), colormap='pink')

plt.subplot(1,2,2)
above_50k["education"].value_counts().plot(kind="bar")

plt.tight_layout(pad=0.3, w_pad=6.0, h_pad=0.9)
```

# Pie Chart for Race Variable

```python
plt.subplot(1,2,1)
below_50k["race"].value_counts().plot(kind="pie", figsize=(14,8))

plt.subplot(1,2,2)
above_50k["race"].value_counts().plot(kind="pie")

plt.tight_layout(pad=0.3, w_pad=6.0, h_pad=0.9)
```

# Horizontal Bar Chart for Marital-Status Variable

```python
plt.subplot(1,2,1)
below_50k["marital-status"].value_counts().plot(kind="barh", figsize=(14,8), colormap='twilight')

plt.subplot(1,2,2)
above_50k["marital-status"].value_counts().plot(kind="barh")

plt.tight_layout(pad=0.3, w_pad=6.0, h_pad=0.9)
```

# Histogram for Number of Hours per Week Variable

```python
plt.subplot(1,2,1)
below_50k["hours-per-week"].plot(kind="hist", figsize=(14,8), colormap='winter')

plt.subplot(1,2,2)
above_50k["hours-per-week"].plot(kind="hist")

plt.tight_layout(pad=0.3, w_pad=6.0, h_pad=0.9)
```

# Area Chart for Native Country Income Variable

```python
below_50k["native-country"].value_counts().plot(kind="pie", figsize=(15,15), colormap='jet', counterclock=False, startangle=90, radius=1800, labels=None)
legend = [str(i)+"-------"+"-------"+str(j) for i, j in zip(below_50k['native-country'].value_counts().keys().tolist(), below_50k['native-country'].value_counts().values.tolist())]
plt.legend(legend, bbox_to_anchor=(1,1))
plt.axis('equal')
plt.title('Native-Countries for <=50k Income Class', fontsize = 25)
plt.show()
```

```python
above_50k["native-country"].value_counts().plot(kind="pie", figsize=(15,15), colormap='hsv', counterclock=False, startangle=90, radius=1800, labels=None)
legend = [str(i)+"-------"+"-------"+str(j) for i, j in zip(above_50k['native-country'].value_counts().keys().tolist(), above_50k['native-country'].value_counts().values.tolist())]
plt.legend(legend, bbox_to_anchor=(1,1))
plt.axis('equal')
plt.title('Native-Countries for >50k Income Class', fontsize = 25)
plt.show()
```
