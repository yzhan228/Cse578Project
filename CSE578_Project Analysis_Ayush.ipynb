```python
df.columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
```

```python
below_50k = df[df["income"] == "<=50K"]
```

```python
below_50k
```

```python
above_50k = df[df["income"] == ">50K"]
```

```python
above_50k
```

```python
below_50k = below_50k[below_50k["age"] != '?']
below_50k = below_50k[below_50k["workclass"] != '?']
below_50k = below_50k[below_50k["fnlwgt"] != '?']
below_50k = below_50k[below_50k["education"] != '?']
below_50k = below_50k[below_50k["education-num"] != '?']
below_50k = below_50k[below_50k["marital-status"] != '?']
below_50k = below_50k[below_50k["occupation"] != '?']
below_50k = below_50k[below_50k["relationship"] != '?']
below_50k = below_50k[below_50k["race"] != '?']
below_50k = below_50k[below_50k["sex"] != '?']
below_50k = below_50k[below_50k["capital-gain"] != '?']
below_50k = below_50k[below_50k["capital-loss"] != '?']
below_50k = below_50k[below_50k["hours-per-week"] != '?']
below_50k = below_50k[below_50k["native-country"] != '?']
below_50k = below_50k[below_50k["income"] != '?']

above_50k = above_50k[above_50k["age"] != '?']
above_50k = above_50k[above_50k["workclass"] != '?']
above_50k = above_50k[above_50k["fnlwgt"] != '?']
above_50k = above_50k[above_50k["education"] != '?']
above_50k = above_50k[above_50k["education-num"] != '?']
above_50k = above_50k[above_50k["marital-status"] != '?']
above_50k = above_50k[above_50k["occupation"] != '?']
above_50k = above_50k[above_50k["relationship"] != '?']
above_50k = above_50k[above_50k["race"] != '?']
above_50k = above_50k[above_50k["sex"] != '?']
above_50k = above_50k[above_50k["capital-gain"] != '?']
above_50k = above_50k[above_50k["capital-loss"] != '?']
above_50k = above_50k[above_50k["hours-per-week"] != '?']
above_50k = above_50k[above_50k["native-country"] != '?']
above_50k = above_50k[above_50k["income"] != '?']
```

# Bar Graph for Sex Variable

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
below_50k["sex"].value_counts().plot(kind="bar", color=['r', 'b'])

plt.subplot(1,2,2)
above_50k["sex"].value_counts().plot(kind='bar', color=['r', 'b'])

plt.tight_layout(pad=0.1, w_pad=7.0, h_pad=0.9)

```

# Histogram for Age Variable

```python
plt.subplot(1,2,1)
below_50k["age"].plot(kind="hist", edgecolor='y', color='b', bins=5)

plt.subplot(1,2,2)
above_50k["age"].plot(kind="hist", edgecolor='black', color='r', bins=5)

plt.tight_layout(pad=0.1, w_pad=7.0, h_pad=0.9)
```

# Bar Chart for Education Type Variable

```python
plt.subplot(1,2,1)
below_50k["education"].value_counts().plot(kind="bar", figsize=(14,8), colormap='pink')

plt.subplot(1,2,2)
above_50k["education"].value_counts().plot(kind="bar")

plt.tight_layout(pad=0.3, w_pad=6.0, h_pad=0.9)
```
