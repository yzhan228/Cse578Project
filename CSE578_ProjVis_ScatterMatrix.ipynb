```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statistics import variance
import math
maindf = pd.read_table("Downloads/adult.data", sep=", ", header=None)
maindf.columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
```

```python
sdf = maindf[['capital-gain', 'education-num', 'income', 'hours-per-week', 'age', 'fnlwgt']]
sdf = sdf.reset_index()
sdf['color'] = np.where(sdf['income'] == '<=50K', 'blue', 'red')
pd.plotting.scatter_matrix(sdf[['capital-gain', 'hours-per-week', 'age', 'education-num']], alpha=0.2, diagonal='kde', c=sdf['color'])
plt.show()
```
